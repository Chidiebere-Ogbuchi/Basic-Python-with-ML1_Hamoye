```python
import pandas as pd

# Specify the URL of the CSV file and the encoding ('ISO-8859-1' or 'latin1')
url = "https://github.com/HamoyeHQ/HDSC-Introduction-to-Python-for-machine-learning/files/7768140/FoodBalanceSheets_E_Africa_NOFLAG.csv"
encoding = "ISO-8859-1"  # or 'latin1'

# Read the CSV file with the specified encoding
FAO = pd.read_csv(url, error_bad_lines=False, encoding=encoding)

# Now, 'FAO' should contain the data from the CSV file without decoding errors
```

```python
##QUESTION 2

# What is the total Protein supply quantity in Madagascar in 2015?

# Filter the data for Madagascar and Protein supply quantity in 2015
madagascar_data = FAO[(FAO['Area'] == 'Madagascar') & (FAO['Element'] == 'Protein supply quantity (g/capita/day)')]

# Get the total protein supply quantity for Madagascar in 2015 ('Y2015' column)
protein_supply_2015 = madagascar_data['Y2015'].sum()

# Now, 'protein_supply_2015' contains the total protein supply quantity in Madagascar in 2015
print("Total Protein Supply Quantity in Madagascar in 2015:", round(protein_supply_2015, 3))
```

```python
# QUESTION 9

# Which year had the least correlation with ‘Element Code’?

# Select the columns of interest, including 'Element Code' and year columns
selected_columns = ['Element Code', 'Y2014', 'Y2015', 'Y2016', 'Y2017', 'Y2018']
selected_data = FAO[selected_columns]

# Calculate the correlation between 'Element Code' and each year column
correlations = selected_data.corr().abs()

# # Exclude 'Element Code' from the correlation matrix
correlations = correlations.drop(['Y2014', 'Y2015', 'Y2016', 'Y2017', 'Y2018'], axis=1)

correlations.sort_values(ascending=True, by='Element Code').iloc[0:1]
```

```python
# QUESTION 13
# What is the total number and percentage of missing data in 2014 to 3 decimal places?

# number of missing values in 'Y2014'
missing_values_2014 = FAO['Y2014'].isna().sum()

# total number of rows in the DataFrame
total_rows = len(FAO)

# percentage of missing data in 2014
percentage_missing_2014 = (missing_values_2014 / total_rows) * 100

# Print the total number and percentage of missing data in 2014 (rounded to 3 decimal places)
print("Total Missing Data in 2014:", missing_values_2014)
print("Percentage of Missing Data in 2014:", round(percentage_missing_2014, 3), "%")
```

```python
# QUESTION 14
# Perform a groupby operation on ‘Element’.  What is the total number of the sum of Processing in 2017?

sel_element_is_processing = FAO[FAO['Element']=='Processing']

sel_element_is_processing.groupby('Element')['Y2017'].sum()
```

```python
# QUESTION 15
# What is the mean and standard deviation across the whole dataset for the year 2017 to 2 decimal places?

Y2017_data = FAO['Y2017']
Y2017_mean = round(Y2017_data.mean(), 2)

Y2017_std = round(Y2017_data.std(), 2)

print(f"for Y2017 the mean is {Y2017_mean} and the standard deviation is {Y2017_std} to 2 dps")
```

```python
# QUESTION 19
# Perform a groupby operation on ‘Element’.  What year has the highest sum of Stock Variation?

selected_columns = ['Element', 'Y2014', 'Y2015', 'Y2016', 'Y2017', 'Y2018']
selected_element_data = FAO[selected_columns]
stock_variation_element = selected_element_data[selected_element_data['Element'] == 'Stock Variation']
sve_grouby = stock_variation_element.groupby('Element').sum().T
sve_grouby.sort_values(by = 'Stock Variation', ascending=False).iloc[0:1]
```
