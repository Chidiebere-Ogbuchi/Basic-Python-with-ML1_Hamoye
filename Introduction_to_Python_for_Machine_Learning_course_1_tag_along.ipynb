```python
#Libraries
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

### Data Wrangling

```python
#import data
url = "https://raw.githubusercontent.com/WalePhenomenon/climate_change/master/fuel_ferc1.csv"
fuel_data = pd.read_csv(url, error_bad_lines=False)
```

```python
fuel_data.head(5)
```

```python
fuel_data.dtypes
```

```python
fuel_data.describe(include = "all")
```

```python
#check for missing values
fuel_data.isnull().sum()
```

```python
#use groupby to count the sum of each unique value in the fuel unit column
fuel_data.groupby('fuel_unit')['fuel_unit'].count()
```

```python
#fill missing value
fuel_data['fuel_unit'] = fuel_data['fuel_unit'].fillna(value='mcf')
```

```python
#check if missing values have been filled
fuel_data.isnull().sum()
```

```python
#group by the report year
fuel_data.groupby('report_year')['report_year'].count()
```

```python
# group by the fuel type code year and print the first entries in all the groups formed
fuel_data.groupby('fuel_type_code_pudl').first()
```

```python
#slice fuel data
fuel_df1 = fuel_data.iloc[0:19000].reset_index(drop=True)
fuel_df2 = fuel_data.iloc[19000:].reset_index(drop=True)
```

```python
#check that the length of both dataframes sum to the expected length
assert len(fuel_data) == (len(fuel_df1) + len(fuel_df2))
```

```python
#an inner merge will lose rows that do not match in both dataframes
pd.merge(fuel_df1, fuel_df2, how = "inner")
```

```python
#outer #removes rows from the right dataframe that do not have a match with the left
#and keeps all rows from the left
pd.merge(fuel_df1, fuel_df2, how = "outer")
```

```python
pd.merge(fuel_df1, fuel_df2, how = "left")
```

```python
#Concatenation is performed with the ​concat() ​function
concat_data = pd.concat([fuel_data, fuel_data]).reset_index(drop=True)

len(concat_data)
```

```python
#check for duplicate rows
concat_data.duplicated().any()
```

### Data Visualization and Representation in Python

```python
#libraries
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

```

```python
# Create a DataFrame

# Assuming you have the fuel_data DataFrame
# Group by 'fuel_unit' and count occurrences, then convert to a DataFrame
fuel_unit_counts_df = pd.DataFrame(fuel_data.groupby('fuel_unit').size().reset_index())

# Rename the columns to avoid conflicts
fuel_unit_counts_df.columns = ['Fuel Unit', 'Count']

# Now, fuel_unit_counts_df is a DataFrame containing the counts of each fuel unit
print(fuel_unit_counts_df)
```

```python

# Create a bar plot
plt.figure(figsize=(7, 4))
plt.xticks(rotation=90)
sns.barplot(data=fuel_unit_counts_df, x='Fuel Unit', y='Count')
plt.xlabel('Fuel Unit')
plt.title('Count of Each Fuel Unit')

# Show the plot
plt.show()
```

```python
# Create the second bar plot with a logarithmic y-scale
plt.figure(figsize=(7, 4))
plt.xticks(rotation=90)
g = sns.barplot(data=fuel_unit_counts_df, x='Fuel Unit', y='Count')
g.set_yscale("log")
g.set_ylim(1, 12000)
plt.xlabel('Fuel Unit')
plt.ylabel('Count (Log Scale)')
plt.title('Count of Each Fuel Unit (Log Scale)')

# Show the second plot
plt.show()
```

```python
# Select a random sample of 50 rows with a fixed random state for reproducibility
sample_df = fuel_data.sample(n=50, random_state=4)

# Create a scatter plot
sns.regplot(x=sample_df["utility_id_ferc1"], y=sample_df["fuel_cost_per_mmbtu"], fit_reg=False)

# Set labels and title
plt.xlabel("Utility ID (ferc1)")
plt.ylabel("Fuel Cost per MMBtu")
plt.title("Scatter Plot of Utility ID vs. Fuel Cost per MMBtu")

# Show the plot
plt.show()
```

```python
# Box Plot
sns.boxplot(x="fuel_type_code_pudl", y="utility_id_ferc1", palette=["m", "g"], data=fuel_data)
plt.title("Box Plot of Utility ID by Fuel Type")
plt.xlabel("Fuel Type")
plt.ylabel("Utility ID")

# Show plot
plt.show()
```

```python

# KDE Plot
sns.kdeplot(sample_df['fuel_cost_per_unit_burned'], shade=True, color="b")
plt.title("KDE Plot of Fuel Cost per Unit Burned")
plt.xlabel("Fuel Cost per Unit Burned")


# Show plot
plt.show()
```
